```python
#plot histograms for all numeric columns in 'dt'
import warnings 
warnings.filterwarnings("ignore")
for i in dt.select_dtypes(include='number').columns:
    sns.histplot(data=dt,x=i)
    plt.show()
```

```python
#plot boxPlot for all numeric columns in 'dt'
import warnings 
warnings.filterwarnings("ignore")
for i in dt.select_dtypes(include='number').columns:
    sns.boxplot(data=dt,x=i)
    plt.show()
```

```python
# Plot scatterplots for multiple columns against 'Life expectancy' to analyze relationships
for i in ['R&D Expenditure (% of GDP)', 'Patents per 100,000 Inhabitants',
       'Unemployment Rate (%)', 'Youth Unemployment Rate (%)',
       'SME Employment (%)', 'Tourism Sector Employment (%)',
       'ICT Sector Employment (%)']:
    sns.scatterplot(data=dt,x='Year',y=i)
    plt.show()
```
